#**Handling missing values**

First we will clean the datasets by handling the missing values. If there are missing values in the Category, Address (country, region...), Phone, Company name fields, we will drop that entry.
I'll start with the google dataset.

```python
old_size = google_dt.shape
google_dt.dropna(subset=['address', 'category', 'country_code', 'city', 'name', 'phone', 'region_code', 'domain', 'zip_code'], inplace=True)
new_size = google_dt.shape
print("Number of Rows(data): ", new_size[0] , "Rows dropped: " ,  old_size[0] - new_size[0]) #Shows the number of entries droped
```

The method above (dropna) checks the columns specified and drops the row if there is a missing value in that column. The inplace specifies if the new dataframe is a copy or the original is modified.

Next we do the same for the other two datasets.

```python
old_size = facebook_dt.shape
facebook_dt.dropna(subset=['address', 'categories', 'country_code', 'city', 'name', 'phone', 'region_code', 'domain', 'zip_code'], inplace=True)
new_size = facebook_dt.shape
print("Number of Rows(data): ", new_size[0] , "Rows dropped: " ,  old_size[0] - new_size[0]) #Shows the number of entries droped
```

```python
old_size = website_dt.shape
website_dt.dropna(subset=['main_city', 'main_country', 'main_region', 'phone', 'legal_name', 'root_domain', 's_category'], inplace=True)
new_size = website_dt.shape
print("Number of Rows(data): ", new_size[0] , "Rows dropped: " ,  old_size[0] - new_size[0]) #Shows the number of entries droped
```

#**Joining the datasets**

I've decided to use the company name the merge key do to it being mostly unique and also because I have experimented using other merge keys and I have seen that using more merge keys will not find the data that is in both datasets and joins them. For example when using the name as the merge key the merged dataset between the facebook and google datasets has 208609 rows meaning that (195757(rows google) + 19496(rows facebook) - 208609(rows merged)) 6644 data was found in the facebook datasets with the same company name as in the google dataset.

```python
g_and_f = pd.merge(google_dt, facebook_dt,  how='outer',  left_on=['name'], right_on=['name'])
g_and_f.shape
```

```python
display(g_and_f)
```

```python
g_and_f.columns.values
```

First I will delete the raw_address and raw_phone columns as they are not needed because we have the normal address and phone columns.

```python
del g_and_f['raw_address']
del g_and_f['raw_phone']
```

As we can see because we used only the name to join we have a lot of column name conflits because a lot of columns have the same name in both datasets. I have decided to keep the columns from the google dataset because it is overall a bigger dataset and the data is cleaner. For the values that are null in the google columns i will copy the data from the facebook columns. For example if adress_x is null i will copy adress_y to it and then will drop adress_y.
I will do this with all conflicting columns.

```python
#Handling address columns
g_and_f.loc[g_and_f["address_x"].isnull(), "address_x"] = g_and_f["address_y"]
del g_and_f['address_y']
g_and_f.rename(columns = {'address_x':'address'}, inplace = True)
```

The loc method just replaces all the values that are null in the adress_x column with the value in the same row of column adress_y.
Now we do this with all conflicting columns.

```python
#Handling category columns
g_and_f.loc[g_and_f["category"].isnull(), "category"] = g_and_f["categories"]
del g_and_f['categories']
```

```python
#Handling city columns
g_and_f.loc[g_and_f["city_x"].isnull(), "city_x"] = g_and_f["city_y"]
del g_and_f['city_y']
g_and_f.rename(columns = {'city_x':'city'}, inplace = True)
```

```python
#Handling country_code columns
g_and_f.loc[g_and_f["country_code_x"].isnull(), "country_code_x"] = g_and_f["country_code_y"]
del g_and_f['country_code_y']
g_and_f.rename(columns = {'country_code_x':'country_code'}, inplace = True)
```

```python
#Handling country_name columns
g_and_f.loc[g_and_f["country_name_x"].isnull(), "country_name_x"] = g_and_f["country_name_y"]
del g_and_f['country_name_y']
g_and_f.rename(columns = {'country_name_x':'country_name'}, inplace = True)
```

```python
#Handling phone columns
g_and_f.loc[g_and_f["phone_x"].isnull(), "phone_x"] = g_and_f["phone_y"]
del g_and_f['phone_y']
g_and_f.rename(columns = {'phone_x':'phone'}, inplace = True)
```

```python
#Handling phone_country_code columns
g_and_f.loc[g_and_f["phone_country_code_x"].isnull(), "phone_country_code_x"] = g_and_f["phone_country_code_y"]
del g_and_f['phone_country_code_y']
g_and_f.rename(columns = {'phone_country_code_x':'phone_country_code'}, inplace = True)
```

```python
#Handling region_code columns
g_and_f.loc[g_and_f["region_code_x"].isnull(), "region_code_x"] = g_and_f["region_code_y"]
del g_and_f['region_code_y']
g_and_f.rename(columns = {'region_code_x':'region_code'}, inplace = True)
```

```python
#Handling region_name columns
g_and_f.loc[g_and_f["region_name_x"].isnull(), "region_name_x"] = g_and_f["region_name_y"]
del g_and_f['region_name_y']
g_and_f.rename(columns = {'region_name_x':'region_name'}, inplace = True)
```

```python
#Handling description columns
g_and_f.loc[g_and_f["text"].isnull(), "text"] = g_and_f["description"]
del g_and_f['description']
g_and_f.rename(columns = {'text':'description'}, inplace = True)
```

```python
#Handling zip_code columns
g_and_f.loc[g_and_f["zip_code_x"].isnull(), "zip_code_x"] = g_and_f["zip_code_y"]
del g_and_f['zip_code_y']
g_and_f.rename(columns = {'zip_code_x':'zip_code'}, inplace = True)
```

```python
#Handling domain columns
g_and_f.loc[g_and_f["domain_x"].isnull(), "domain_x"] = g_and_f["domain_y"]
del g_and_f['domain_y']
g_and_f.rename(columns = {'domain_x':'domain'}, inplace = True)
```

Now that we are done with the columns i will check wether there are any missing values in the import columns (address, category, country_code, city, name,phone, region_code, domain, zip_code)

```python
print(g_and_f.isnull().any())
```

Our most important columns dont have any missing values so everything looks good.

Next we will do the final merge with the g_and_f dataset and the website_dt using the same principle and also using the company name to merge.

(195757(rows google) + 19496(rows facebook) + 27944(rows website) - 232118(rows merged_all))

11079 data rows with the same company name in all three datasets.

```python
Datasets_joined = pd.merge(g_and_f, website_dt,  how='outer',  left_on=['name'], right_on=['legal_name'])
Datasets_joined.shape
```

```python
display(Datasets_joined)
```

```python
Datasets_joined.columns.values
```

Now we do the same thing we did before but we also have to do it for the name column because for the company names unique to the website dataset we will need to copy them to the name column and the legal_name column should be droped.

```python
#Handling name columns
Datasets_joined.loc[Datasets_joined["name"].isnull(), "name"] = Datasets_joined["legal_name"]
del Datasets_joined['legal_name']
```

```python
#Handling category columns
Datasets_joined.loc[Datasets_joined["category"].isnull(), "category"] = Datasets_joined["s_category"]
del Datasets_joined['s_category']
```

```python
#Handling city columns
Datasets_joined.loc[Datasets_joined["city"].isnull(), "city"] = Datasets_joined["main_city"]
del Datasets_joined['main_city']
```

```python
#Handling country_name columns
Datasets_joined.loc[Datasets_joined["country_name"].isnull(), "country_name"] = Datasets_joined["main_country"]
del Datasets_joined['main_country']
```

```python
#Handling region_name columns
Datasets_joined.loc[Datasets_joined["region_name"].isnull(), "region_name"] = Datasets_joined["main_region"]
del Datasets_joined['main_region']
```

```python
#Handling phone columns
Datasets_joined.loc[Datasets_joined["phone_x"].isnull(), "phone_x"] = Datasets_joined["phone_y"]
del Datasets_joined['phone_y']
Datasets_joined.rename(columns = {'phone_x':'phone'}, inplace = True)
```

```python
#Handling domain columns
Datasets_joined.loc[Datasets_joined["domain"].isnull(), "domain"] = Datasets_joined["root_domain"]
del Datasets_joined['root_domain']
```

Now we can check for the final dataset if our important data has missing values.

```python
print(Datasets_joined.isnull().any())
```

As we can see there are null values in our addresses column because the website dataset does not have any address. But our Category, City, Country, Name, Phone, Region and Domain do not have any null values.
